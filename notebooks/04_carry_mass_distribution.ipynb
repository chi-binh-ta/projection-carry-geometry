# 04 - Carry mass distribution

This notebook samples random decimal products and visualizes the distribution of carry mass. It also compares carry mass with carry depth and carry entropy.

```python
from pathlib import Path
import sys

sys.path.append("../src")
if (Path.cwd() / "src").exists():
    sys.path.append(str(Path.cwd() / "src"))
elif (Path.cwd().parent / "src").exists():
    sys.path.append(str(Path.cwd().parent / "src"))

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from experiments import random_digit_number
from metrics import projection_summary
```

```python
rng = np.random.default_rng(20260504)
rows = []
for trial in range(120):
    x = random_digit_number(num_digits=10, B=10, density=1.0, rng=rng)
    y = random_digit_number(num_digits=10, B=10, density=1.0, rng=rng)
    summary = projection_summary(x, y, 10)
    rows.append({
        "trial": trial,
        "carry_mass": summary["carry_mass"],
        "carry_depth": summary["carry_depth"],
        "carry_entropy": summary["carry_entropy"],
        "max_raw_coefficient": summary["max_raw_coefficient"],
    })

frame = pd.DataFrame(rows)
frame.describe()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(13, 3.8))
frame["carry_mass"].hist(ax=axes[0], bins=20)
axes[0].set_title("carry_mass")
frame["carry_depth"].hist(ax=axes[1], bins=range(0, int(frame["carry_depth"].max()) + 2))
axes[1].set_title("carry_depth")
frame["carry_entropy"].hist(ax=axes[2], bins=20)
axes[2].set_title("carry_entropy")
for ax in axes:
    ax.grid(alpha=0.25)
plt.tight_layout()
plt.show()
```

`carry_mass` measures total transported carry. `carry_depth` records how far carries appear along the degree axis. `carry_entropy` is high when positive carries are spread across many positions rather than concentrated in one place.
