# Empirical Scaling Laws

Hypothesis: support-sumset geometry controls carry complexity through diagonal representation concentration. In particular, additive energy and maximum representation count should predict carry mass, while sumset size and density help explain carry count and cascade behavior.

Geometry predictors are not full determinants. Support geometry describes potential diagonal concentration through `rho_k` and additive energy, but digit amplitudes and incoming carry decide the realized local threshold events. This is why the empirical relation is predictive rather than exact.
