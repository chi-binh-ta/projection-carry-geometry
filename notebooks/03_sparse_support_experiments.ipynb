# 03 - Sparse support experiments

Sparse digit support means many digit positions are zero. This notebook compares low and high support density and records how that changes carry metrics.

```python
from pathlib import Path
import sys

sys.path.append("../src")
if (Path.cwd() / "src").exists():
    sys.path.append(str(Path.cwd() / "src"))
elif (Path.cwd().parent / "src").exists():
    sys.path.append(str(Path.cwd().parent / "src"))

import matplotlib.pyplot as plt
from experiments import experiment_sparse_support
```

```python
FIGURES = Path("figures") if (Path.cwd() / "figures").exists() else Path("../figures")
frame = experiment_sparse_support(
    n_trials=40,
    num_digits=10,
    B=10,
    densities=[0.1, 0.3, 0.7, 1.0],
    output_csv=str(FIGURES / "sparse_support_notebook_metrics.csv"),
)
frame.head()
```

```python
summary = frame.groupby("density")[[
    "carry_mass",
    "carry_depth",
    "carry_entropy",
    "x_support_size",
    "y_support_size",
    "raw_support_size",
]].mean()
summary
```

```python
ax = summary[["carry_mass", "raw_support_size"]].plot(marker="o", figsize=(7, 4))
ax.set_title("Sparse support: low density vs high density")
ax.set_xlabel("density")
ax.grid(alpha=0.25)
plt.show()
```

Lower density usually shrinks the support sumset and reduces the amount of raw coefficient mass that must be normalized by carries. Dense supports create more collisions on anti-diagonals, which often increases carry mass.
