# 05 - Base dependence

This notebook compares base `B = 2, 10, 16, 100` using the same number of digits. Larger bases allow larger final digits, but raw coefficients can also become larger because individual digit products are larger.

```python
from pathlib import Path
import sys

sys.path.append("../src")
if (Path.cwd() / "src").exists():
    sys.path.append(str(Path.cwd() / "src"))
elif (Path.cwd().parent / "src").exists():
    sys.path.append(str(Path.cwd().parent / "src"))

import matplotlib.pyplot as plt
from experiments import experiment_base_dependence
```

```python
FIGURES = Path("figures") if (Path.cwd() / "figures").exists() else Path("../figures")
frame = experiment_base_dependence(
    n_trials=40,
    num_digits=10,
    bases=[2, 10, 16, 100],
    output_csv=str(FIGURES / "base_dependence_notebook_metrics.csv"),
)
frame.head()
```

```python
summary = frame.groupby("B")[[
    "carry_mass",
    "carry_depth",
    "carry_entropy",
    "max_raw_coefficient",
]].mean()
summary
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 3.8))
summary[["carry_mass", "max_raw_coefficient"]].plot(kind="bar", ax=axes[0])
axes[0].set_title("Mass and raw coefficient scale")
summary[["carry_depth", "carry_entropy"]].plot(kind="bar", ax=axes[1])
axes[1].set_title("Carry shape metrics")
for ax in axes:
    ax.grid(axis="y", alpha=0.25)
plt.tight_layout()
plt.show()
```

Base dependence has two competing effects: higher bases increase the threshold for creating carry, while also increasing possible digit-product magnitudes. These metrics make that tradeoff visible.
