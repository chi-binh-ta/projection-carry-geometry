# Carry as Nonlinear Mass Flow

After anti-diagonal projection, the raw diagonal coefficients `c_k` can be viewed as mass on the degree axis. Carry normalization transfers excess mass from degree `k` to degree `k + 1` while keeping a valid base-`B` digit at degree `k`.

The carry recurrence is:

$$s_k = c_k + r_k$$

$$d_k = s_k \bmod B$$

$$r_{k+1} = \left\lfloor \frac{s_k}{B} \right\rfloor$$

Here `r_k` is incoming carry, `d_k` is the normalized digit kept at degree `k`, and `r_{k+1}` is outgoing mass transported along the directed edge `k -> k + 1`.

## Demo: 99 x 99

`99 * 99` creates a more visibly carry-heavy decimal profile.

```python
out_99 = multiply_projection_carry(99, 99, 10)
flow_99 = carry_flow_table(out_99["raw_diagonal_sums"], 10)
print("raw_diagonal_sums:", out_99["raw_diagonal_sums"])
print("final_digits:", out_99["final_digits"])
print("result:", out_99["result"])
flow_99
```

```python
d_99, _ = carry_normalize(out_99["raw_diagonal_sums"], 10)
carry_value_preservation(out_99["raw_diagonal_sums"], d_99, 10)
```

Carry normalization preserves numerical value because each step rewrites `c_k + r_k` as `d_k + B r_{k+1}`. After multiplying by `B^k` and summing over `k`, the carry terms telescope.
