# Paper Result Consolidation

This notebook freezes the paper-facing outputs: raw correlations, normalized correlations, grouped scaling summaries, and paper figures.

```python
from pathlib import Path
import sys

sys.path.append("../src")
if (Path.cwd() / "src").exists():
    sys.path.append(str(Path.cwd() / "src"))
elif (Path.cwd().parent / "src").exists():
    sys.path.append(str(Path.cwd().parent / "src"))

import pandas as pd
from IPython.display import Image, display
from paper_figures import make_normalized_scaling_figures, make_scaling_figures
from paper_tables import make_correlation_tables
from scaling_experiments import correlation_table, normalized_correlation_table, run_scaling_grid, summarize_scaling_results
```

```python
RESULTS = Path("results") if (Path.cwd() / "results").exists() else Path("../results")
PAPER = Path("paper") if (Path.cwd() / "paper").exists() else Path("../paper")
RESULTS.mkdir(parents=True, exist_ok=True)
(PAPER / "figures").mkdir(parents=True, exist_ok=True)
(PAPER / "tables").mkdir(parents=True, exist_ok=True)

scaling_csv = RESULTS / "scaling_grid.csv"
if scaling_csv.exists():
    data = pd.read_csv(scaling_csv)
else:
    data = run_scaling_grid(
        output_csv=str(scaling_csv),
        bases=[2, 10, 16],
        num_digits_list=[8, 16],
        densities=[0.2, 0.5, 0.7, 1.0],
        n_trials=20,
    )

data.head()
```

```python
make_correlation_tables(str(scaling_csv), str(PAPER / "tables"))
raw_corr = correlation_table(data)
norm_corr = normalized_correlation_table(data)
summary = summarize_scaling_results(data)

raw_corr
```

```python
make_scaling_figures(str(scaling_csv), str(PAPER / "figures"))
make_normalized_scaling_figures(str(scaling_csv), str(PAPER / "figures"))

figure_names = [
    "additive_energy_vs_carry_mass.png",
    "max_representation_vs_carry_mass.png",
    "size_sumset_vs_carry_count.png",
    "normalized_additive_energy_vs_carry_mass_per_digit.png",
    "density_vs_cascade_fraction.png",
    "base_vs_carry_mass_per_digit.png",
    "num_digits_vs_carry_count_per_digit.png",
]
for name in figure_names:
    display(Image(filename=str(PAPER / "figures" / name)))
```

Interpretation: support-sumset geometry is a structural predictor of carry complexity. Additive energy and representation concentration describe potential diagonal collisions, while normalized carry metrics control for digit length and support size. The remaining variation is expected because digit amplitudes and incoming carry determine realized threshold crossings.
