# 01 - 123 times 321 demo

This notebook demonstrates Projection--Carry Geometry on `123 * 321` in base 10.

The central identity is:

**multiplication = outer product + anti-diagonal projection + carry**.

The raw diagonal sums are `[3, 8, 14, 8, 3]`. The coefficient `14` is not a base-10 digit, so carry normalization sends one unit of mass to the next degree and leaves digit `4` behind.
